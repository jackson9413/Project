```python
# Method 2: Percentage of "Yes" responses by college (for colleges with at least 10 students)
# Calculate percentage of "Yes" responses for each college
college_stats = data.groupby('College_Name').agg({
    'Willing_to_Pay_for_Access': ['count', lambda x: (x == 'Yes').sum()]
}).round(2)

# Flatten column names
college_stats.columns = ['total_students', 'yes_count']

# Calculate percentage
college_stats['yes_percentage'] = (college_stats['yes_count'] / college_stats['total_students'] * 100).round(1)

# Filter colleges with at least 10 students and get top 15 by percentage
colleges_filtered = college_stats[college_stats['total_students'] >= 10]
top_colleges_percentage = colleges_filtered.sort_values('yes_percentage', ascending=False).head(15)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_colleges_percentage['yes_percentage'], y=top_colleges_percentage.index, palette='plasma')
plt.title('Top 15 Colleges/Universities by Percentage of "Yes" Responses\n(Willing to Pay for Access - Min 10 students)')
plt.xlabel('Percentage of "Yes" Responses (%)')
plt.ylabel('College/University Name')
plt.tight_layout()
plt.show()

print("Top 15 colleges by percentage of 'Yes' responses (min 10 students):")
for i, (college, row) in enumerate(top_colleges_percentage.iterrows(), 1):
    print(f"{i}. {college}: {row['yes_percentage']:.1f}% ({row['yes_count']}/{row['total_students']} students)")
```

## Daily_Usage_Hours

```python
# Method 2: Percentage of "Yes" responses by daily usage hours (binned into ranges)
# Since Daily_Usage_Hours is continuous, we need to bin it into categories

# First, let's check the range of daily usage hours
print("Daily Usage Hours Statistics:")
print(f"Min: {data['Daily_Usage_Hours'].min()}")
print(f"Max: {data['Daily_Usage_Hours'].max()}")
print(f"Mean: {data['Daily_Usage_Hours'].mean():.2f}")
print(f"Median: {data['Daily_Usage_Hours'].median():.2f}")

# Create bins for daily usage hours
# We'll create reasonable ranges based on the data distribution
bins = [0, 1, 2, 3, 4, 5, float('inf')]
labels = ['0-1 hours', '1-2 hours', '2-3 hours', '3-4 hours', '4-5 hours', '5+ hours']

# Create a new column with binned usage hours
data_temp = data.copy()
data_temp['Usage_Hours_Binned'] = pd.cut(data_temp['Daily_Usage_Hours'], bins=bins, labels=labels, include_lowest=True)

# Calculate percentage of "Yes" responses for each usage hour range
usage_stats = data_temp.groupby('Usage_Hours_Binned').agg({
    'Willing_to_Pay_for_Access': ['count', lambda x: (x == 'Yes').sum()]
}).round(2)

# Flatten column names
usage_stats.columns = ['total_students', 'yes_count']

# Calculate percentage
usage_stats['yes_percentage'] = (usage_stats['yes_count'] / usage_stats['total_students'] * 100).round(1)

# Filter bins with at least 10 students
usage_filtered = usage_stats[usage_stats['total_students'] >= 10]

plt.figure(figsize=(12, 6))
sns.barplot(x=usage_filtered.index, y=usage_filtered['yes_percentage'], palette='viridis')
plt.title('Percentage of "Yes" Responses by Daily Usage Hours\n(Willing to Pay for Access - Min 10 students per bin)')
plt.xlabel('Daily Usage Hours Range')
plt.ylabel('Percentage of "Yes" Responses (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Percentage of 'Yes' responses by daily usage hours (min 10 students per bin):")
for usage_range, row in usage_filtered.iterrows():
    print(f"{usage_range}: {row['yes_percentage']:.1f}% ({row['yes_count']}/{row['total_students']} students)")
```

```python
# Method 2: Internet_Access - Percentage of "Yes" responses by internet access type
internet_stats = data.groupby('Internet_Access')['Willing_to_Pay_for_Access'].agg([
    'count',
    lambda x: (x == 'Yes').sum(),
    lambda x: (x == 'Yes').mean() * 100
]).round(2)

internet_stats.columns = ['total_count', 'yes_count', 'yes_percentage']

# Filter for groups with at least 10 students
internet_filtered = internet_stats[internet_stats['total_count'] >= 10]

# Sort by percentage descending
internet_sorted = internet_filtered.sort_values('yes_percentage', ascending=False)

print("Internet_Access - Willingness to Pay Analysis (Method 2)")
print("=" * 60)
print(f"{'Internet Access Type':<25} {'Yes %':<8} {'Yes Count':<10} {'Total':<8}")
print("-" * 60)

for access_type, row in internet_sorted.iterrows():
    print(f"{access_type:<25} {row['yes_percentage']:>6.1f}% {row['yes_count']:>8.0f} {row['total_count']:>10.0f}")

# Create bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(internet_sorted)), internet_sorted['yes_percentage'], 
               color=['#2E8B57', '#4682B4', '#DAA520', '#DC143C', '#8A2BE2'][:len(internet_sorted)])

plt.title('Willingness to Pay for AI Tools by Internet Access Type\n(Method 2: Percentage of "Yes" Responses)', 
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Internet Access Type', fontsize=12)
plt.ylabel('Percentage Willing to Pay (%)', fontsize=12)

# Customize x-axis labels
plt.xticks(range(len(internet_sorted)), internet_sorted.index, rotation=45, ha='right')

# Add percentage labels on bars
for i, (access_type, row) in enumerate(internet_sorted.iterrows()):
    plt.text(i, row['yes_percentage'] + 1, f"{row['yes_percentage']:.1f}%", 
             ha='center', va='bottom', fontweight='bold')

plt.ylim(0, max(internet_sorted['yes_percentage']) * 1.15)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Summary statistics and insights
print(f"\nSummary Statistics:")
print(f"Total internet access types analyzed: {len(internet_sorted)}")
print(f"Internet access types excluded (< 10 students): {len(internet_stats) - len(internet_sorted)}")

if len(internet_sorted) > 0:
    highest_access = internet_sorted.iloc[0]
    lowest_access = internet_sorted.iloc[-1]
    print(f"Highest willingness to pay: {internet_sorted.index[0]} ({highest_access['yes_percentage']:.1f}%)")
    print(f"Lowest willingness to pay: {internet_sorted.index[-1]} ({lowest_access['yes_percentage']:.1f}%)")
    
    # Calculate difference between highest and lowest
    if len(internet_sorted) > 1:
        percentage_difference = highest_access['yes_percentage'] - lowest_access['yes_percentage']
        print(f"Difference in willingness to pay: {percentage_difference:.1f} percentage points")
        
        # Specific insights about different internet access types
        print(f"\nInternet Access Insights:")
        for access_type, row in internet_sorted.iterrows():
            print(f"- {access_type}: {row['yes_percentage']:.1f}% willing to pay ({row['yes_count']:.0f}/{row['total_count']:.0f} students)")
        
        # Identify premium vs basic internet access patterns
        if len(internet_sorted) >= 2:
            avg_percentage = internet_sorted['yes_percentage'].mean()
            above_avg = internet_sorted[internet_sorted['yes_percentage'] > avg_percentage]
            below_avg = internet_sorted[internet_sorted['yes_percentage'] <= avg_percentage]
            
            print(f"\nAverage willingness to pay across all internet types: {avg_percentage:.1f}%")
            if len(above_avg) > 0:
                print(f"Above average: {', '.join(above_avg.index)}")
            if len(below_avg) > 0:
                print(f"Below average: {', '.join(below_avg.index)}")
                
        # Check for correlation with internet access quality
        print(f"\nInternet Access Quality Analysis:")
        high_quality_types = ['Fiber', 'Broadband', 'Cable', 'High-speed']
        low_quality_types = ['Dial-up', 'Satellite', 'Mobile data', 'Limited']
        
        high_quality_data = internet_sorted[internet_sorted.index.str.contains('|'.join(high_quality_types), case=False, na=False)]
        low_quality_data = internet_sorted[internet_sorted.index.str.contains('|'.join(low_quality_types), case=False, na=False)]
        
        if len(high_quality_data) > 0:
            high_avg = high_quality_data['yes_percentage'].mean()
            print(f"High-quality internet access average willingness: {high_avg:.1f}%")
        
        if len(low_quality_data) > 0:
            low_avg = low_quality_data['yes_percentage'].mean()
            print(f"Lower-quality internet access average willingness: {low_avg:.1f}%")
            
        if len(high_quality_data) > 0 and len(low_quality_data) > 0:
            quality_diff = high_avg - low_avg
            print(f"Quality difference impact: {quality_diff:.1f} percentage points")
```
