```python
# Method 2: Percentage of "Yes" responses by college (for colleges with at least 10 students)
# Calculate percentage of "Yes" responses for each college
college_stats = data.groupby('College_Name').agg({
    'Willing_to_Pay_for_Access': ['count', lambda x: (x == 'Yes').sum()]
}).round(2)

# Flatten column names
college_stats.columns = ['total_students', 'yes_count']

# Calculate percentage
college_stats['yes_percentage'] = (college_stats['yes_count'] / college_stats['total_students'] * 100).round(1)

# Filter colleges with at least 10 students and get top 15 by percentage
colleges_filtered = college_stats[college_stats['total_students'] >= 10]
top_colleges_percentage = colleges_filtered.sort_values('yes_percentage', ascending=False).head(15)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_colleges_percentage['yes_percentage'], y=top_colleges_percentage.index, palette='plasma')
plt.title('Top 15 Colleges/Universities by Percentage of "Yes" Responses\n(Willing to Pay for Access - Min 10 students)')
plt.xlabel('Percentage of "Yes" Responses (%)')
plt.ylabel('College/University Name')
plt.tight_layout()
plt.show()

print("Top 15 colleges by percentage of 'Yes' responses (min 10 students):")
for i, (college, row) in enumerate(top_colleges_percentage.iterrows(), 1):
    print(f"{i}. {college}: {row['yes_percentage']:.1f}% ({row['yes_count']}/{row['total_students']} students)")
```

## Daily_Usage_Hours

```python
# Method 2: Percentage of "Yes" responses by daily usage hours (binned into ranges)
# Since Daily_Usage_Hours is continuous, we need to bin it into categories

# First, let's check the range of daily usage hours
print("Daily Usage Hours Statistics:")
print(f"Min: {data['Daily_Usage_Hours'].min()}")
print(f"Max: {data['Daily_Usage_Hours'].max()}")
print(f"Mean: {data['Daily_Usage_Hours'].mean():.2f}")
print(f"Median: {data['Daily_Usage_Hours'].median():.2f}")

# Create bins for daily usage hours
# We'll create reasonable ranges based on the data distribution
bins = [0, 1, 2, 3, 4, 5, float('inf')]
labels = ['0-1 hours', '1-2 hours', '2-3 hours', '3-4 hours', '4-5 hours', '5+ hours']

# Create a new column with binned usage hours
data_temp = data.copy()
data_temp['Usage_Hours_Binned'] = pd.cut(data_temp['Daily_Usage_Hours'], bins=bins, labels=labels, include_lowest=True)

# Calculate percentage of "Yes" responses for each usage hour range
usage_stats = data_temp.groupby('Usage_Hours_Binned').agg({
    'Willing_to_Pay_for_Access': ['count', lambda x: (x == 'Yes').sum()]
}).round(2)

# Flatten column names
usage_stats.columns = ['total_students', 'yes_count']

# Calculate percentage
usage_stats['yes_percentage'] = (usage_stats['yes_count'] / usage_stats['total_students'] * 100).round(1)

# Filter bins with at least 10 students
usage_filtered = usage_stats[usage_stats['total_students'] >= 10]

plt.figure(figsize=(12, 6))
sns.barplot(x=usage_filtered.index, y=usage_filtered['yes_percentage'], palette='viridis')
plt.title('Percentage of "Yes" Responses by Daily Usage Hours\n(Willing to Pay for Access - Min 10 students per bin)')
plt.xlabel('Daily Usage Hours Range')
plt.ylabel('Percentage of "Yes" Responses (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Percentage of 'Yes' responses by daily usage hours (min 10 students per bin):")
for usage_range, row in usage_filtered.iterrows():
    print(f"{usage_range}: {row['yes_percentage']:.1f}% ({row['yes_count']}/{row['total_students']} students)")
```
