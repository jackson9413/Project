# Multiple Variables

## Daily_Usage_Hours

Use "Daily_Usage_Hours" as a target variable while other columns as the predictors.

### Daily_Usage_Hours ~ College_Name

```python
# group by College_Name and calculate mean Daily_Usage_Hours
college_usage = data.groupby('College_Name')['Daily_Usage_Hours'].mean().sort_values(ascending=False).head(15)
plt.figure(figsize=(12, 6))
sns.barplot(x=college_usage.values, y=college_usage.index, palette='magma')
plt.title('Top 10 Colleges by Average Daily Usage Hours of AI Tools')
plt.xlabel('Average Daily Usage Hours')
plt.ylabel('College Name')
plt.xticks(rotation=45)
plt.show()
```

```python
# draw a box plot for Daily_Usage_Hours by College_Name
# order by mean Daily_Usage_Hours 
# choose top 15 colleges by median Daily_Usage_Hours
top_colleges = data.groupby('College_Name')['Daily_Usage_Hours'].mean().sort_values(ascending=False).head(15).index
plt.figure(figsize=(12, 6))
sns.boxplot(x='Daily_Usage_Hours', y='College_Name', data=data[data['College_Name'].isin(top_colleges)], order=top_colleges, palette='Set3')
plt.title('Daily Usage Hours of AI Tools by Top 15 Colleges')
plt.xlabel('Daily Usage Hours')
plt.ylabel('College Name') 
plt.show()
```

Not huge difference.

### Daily_Usage_Hours ~ Stream

```python
# group by Stream and calculate mean Daily_Usage_Hours
college_usage = data.groupby('Stream')['Daily_Usage_Hours'].mean().sort_values(ascending=False).head(15)
plt.figure(figsize=(12, 6))
sns.barplot(x=college_usage.values, y=college_usage.index, palette='magma')
plt.title('Top 10 Streams by Average Daily Usage Hours of AI Tools')
plt.xlabel('Average Daily Usage Hours')
plt.ylabel('Stream Name')
plt.xticks(rotation=45)
plt.show()
```

```python
# draw a box plot for Daily_Usage_Hours by Stream
# order by mean Daily_Usage_Hours
top_streams = data.groupby('Stream')['Daily_Usage_Hours'].mean().sort_values(ascending=False).head(15).index
plt.figure(figsize=(12, 6))
sns.boxplot(x='Stream', y='Daily_Usage_Hours', data=data[data['Stream'].isin(top_streams)], order=top_streams, palette='Set3')
plt.title('Daily Usage Hours of AI Tools by Top 15 Streams')
plt.xlabel('Stream')
plt.ylabel('Daily_Usage_Hours')
plt.show()
```
